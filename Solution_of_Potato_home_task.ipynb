```python
#Firstly we should to import the numpy and pandas libraries.....
import numpy as np
import pandas as pd
```

```python
#Import the data from this file
data = pd.read_csv(r'C:\Users\shiva\Downloads\correct_twitter_201904.tsv',sep='\t')
```

```python
data.head(5)
```

```python
#create dataframe...
tweets_df = pd.DataFrame(data)
tweets_df
```

```python
#According to question i take music term....
term = "music"
```

```python
#I create the variable "filtered_tweets" and This method is applied to the text column and checks each tweet's content to see if it contains the substring defined in term.
filtered_tweets = tweets_df[tweets_df['text'].str.contains(term, case=False)]
filtered_tweets 
```

```python
tweets_per_day = filtered_tweets['created_at']
```

```python
type(tweets_per_day)
```

```python
#Question 1) How many tweets were posted containing the term on each day?
pd.to_datetime(tweets_per_day).dt.date.value_counts().sort_index()
```

```python
 # Question 2) How many unique users posted a tweet containing the term?
 unique_users = filtered_tweets['author_handle'].nunique()
 unique_users
```

```python
#Question 3) How many likes did tweets containing the term get, on average?
average_likes = filtered_tweets['like_count'].mean()
average_likes
```

```python
#Question 4) Where (in terms of place IDs) did the tweets come from?
place_ids = filtered_tweets['place_id'].unique()
place_ids
```

```python
#Question 5) What times of day were the tweets posted at? 
times_of_day = filtered_tweets['created_at']
pd.to_datetime(times_of_day).dt.time.value_counts()
```

```python
# Question 6) Which user posted the most tweets containing the term?
author_most_tweets = filtered_tweets['author_handle'].value_counts().idxmax()
print(author_most_tweets)
```
